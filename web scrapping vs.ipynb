```python
import requests
from bs4 import BeautifulSoup
```

```python
import pandas as ps
```

```python
url=('https://www.flipkart.com/search?q=cpu+hp+core+i5+windows+10&sid=6bo%2Cnl4%2Cdze&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=RECENT&suggestionId=cpu+hp+core+i5+windows+10%7CTower+PCs&requestId=cd07b8a0-cfdc-4f37-a54e-d2a8024982d6&as-searchtext=cpu')
```

```python
res=requests.get(url)
```

```python
res
```

```python
soup=BeautifulSoup(res.text)
```

```python
soup
```

```python
cpu_names=soup.select(".s1Q9rs")
```

```python
cpu_names
```

```python
cpu_prices=soup.select("._8VNy32 ._30jeq3")
```

```python
cpu_prices
```

```python
cpu_offers=soup.select("._8VNy32 span")
```

```python
cpu_offers
```

```python
df=ps.DataFrame(cpu_names,columns=['cpu_names'])
```

```python
df
```

```python
df['cpu_prices']=cpu_prices
```

```python
df
```

```python
df['cpu_offers']=cpu_offers
```

```python
df
```

```python
df.to_csv('vs.csv')
```
